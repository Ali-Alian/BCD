```python
# Loud the data folder and the images
col_names = ['REFNUM', 'BG', 'CLASS', 'SEVERITY', 'X', 'Y', 'RADIUS']
df = pd.read_csv('data2.txt', sep="\s+", names=col_names, header=None)
df['CANCER'] = df['SEVERITY'].apply(lambda x: 1 if x in ['B', 'M'] else 0)

images_path = "all-mias"

all_images = []
all_labels = []
all_groups = []

for filename in sorted(os.listdir(images_path)):
    if filename.lower().endswith('.pgm'):
        ref_num = os.path.splitext(filename)[0]
        record = df[df['REFNUM'] == ref_num]

        if not record.empty:
            full_path = os.path.join(images_path, filename)
            img_array = cv2.imread(full_path, cv2.IMREAD_GRAYSCALE)
            
            clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
            img_eq = clahe.apply(img_array)

            labels = record['CANCER'].iloc[0]
            x, y, radius = record[['X', 'Y', 'RADIUS']].iloc[0]
            
            # Handle ROI extraction
            if labels == 1 and pd.notna(x) and pd.notna(y) and pd.notna(radius):
                # Adjust for bottom-left origin: flip Y coordinate
                x, y, radius = int(x), int(1024 - float(y)), int(radius)  # Y = 1024 - y for top-left origin
                # Crop a square ROI around (x, y) with size 2*radius
                roi = img_eq[max(0, y-radius):min(1024, y+radius), max(0, x-radius):min(1024, x+radius)]
                # Ensure ROI is not empty; resize to a fixed size (e.g., 128x128) for consistency
                if roi.size > 0:
                    roi = cv2.resize(roi, (256, 256), interpolation=cv2.INTER_AREA)
                else:
                    roi = img_eq  # Fallback to full image if ROI is invalid
            else:
                # For normal images or missing coordinates using all the image in full size
                roi = cv2.resize(img_eq, (256, 256), interpolation=cv2.INTER_AREA)
            
            all_images.append(roi)
            all_labels.append(labels)
            all_groups.append(ref_num[:-1])

print("Data has been louded successfully")
# plt.imshow(img_eq, cmap='gray')
## Print out the table from dataset
print(f"Image list {len(all_images)}")
print(f"Labels list {len(all_labels)}")
df.head(5)
# full details of the table
```

# GLCM

#### 2 - applyed the Gray-Level Co-occurrence Matrix (GLCM)


```python
def extract_glcm_features(patch, 
                          distances=[1, 3, 5], 
                          angles=[0, np.pi/4, np.pi/2, 3*np.pi/4], 
                          levels=256):
    """
    Compute GLCM and summary stats for a single grayscale patch (2D ndarray).
    Returns a dict: e.g. {'contrast_mean':, 'contrast_var':, 'homogeneity_mean':, }
    """
    patch = (patch * (levels - 1)).astype(np.uint8) if patch.max() <= 1 else patch.astype(np.uint8)
    
    glcm = graycomatrix(patch,
                        distances=distances,
                        angles=angles,
                        levels=levels,
                        symmetric=True,
                        normed=True)

    props = ['contrast', 'dissimilarity', 'homogeneity', 
             'ASM', 'energy', 'correlation']
    feats = {}
    for prop in props:
        mat = graycoprops(glcm, prop)  # shape = (len(distances), len(angles))
        feats[f'{prop}_mean'] = mat.mean()
        feats[f'{prop}_var']  = mat.var()
    return feats
```

```python
import numpy as np
import pandas as pd
import cv2
from skimage.feature import local_binary_pattern
import pickle

from sklearn.model_selection import GridSearchCV, GroupKFold, GroupShuffleSplit
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

def main():
    """
    Main function to run the GridSearchCV and GroupKFold pipeline.
    """
    # --- 1. Load Pre-extracted Data ---
    # This section assumes you have already run the initial data preparation 
    # and saved the features and ROIs, as in your original LBP_Final notebook.
    try:
        with open('data/LBP_GLCM_FEATURES.pkl', 'rb') as f:
            final_df = pickle.load(f)
        with open('data/ROIS.pkl', 'rb') as f:
            ROIS = pickle.load(f)
    except FileNotFoundError:
        print("Error: Feature files not found. Please run the initial feature extraction")
        print("and data saving cells from your LBP_Final.ipynb notebook first.")
        # Creating dummy data for demonstration if files are not found
        print("Creating dummy data to proceed with the demonstration...")
        num_samples = 300
        data = {
            'LBP_0': np.random.rand(num_samples),
            'LBP_1': np.random.rand(num_samples),
            'GLCM_contrast_mean': np.random.rand(num_samples),
            'label': np.random.randint(0, 2, num_samples),
            'refnum': np.random.randint(1, 100, num_samples) # Simulating patient IDs
        }
        final_df = pd.DataFrame(data)
    
    print("Data loaded successfully.")
    print(f"Dataset shape: {final_df.shape}")

    # --- 2. Prepare Data for Modeling ---
    # The 'refnum' column will be used as the group identifier for patient-aware splitting.
    X = final_df.drop(['label', 'refnum'], axis=1, errors='ignore')
    y = final_df['label']
    groups = final_df['refnum']

    # --- 3. Split Data into Training and Held-Out Test Sets (Patient-Aware) ---
    print("Splitting data into training and test sets (80/20 split of patients)...")
    gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
    train_idx, test_idx = next(gss.split(X, y, groups))
    
    X_tr, X_te = X.iloc[train_idx], X.iloc[test_idx]
    y_tr, y_te = y.iloc[train_idx], y.iloc[test_idx]
    groups_tr = groups.iloc[train_idx]
    
    print(f"Training set size: {len(X_tr)} images")
    print(f"Test set size: {len(X_te)} images")

    # --- 4. Define Cross-Validation Strategy and Models ---
    NFOLDS = 5
    gkf = GroupKFold(n_splits=NFOLDS)
    print(f"Cross-validation strategy: {NFOLDS}-Fold GroupKFold")

    # Define pipelines for scaling and classification
    pipe_rf = Pipeline([
        ('scaler', StandardScaler()),
        ('clf', RandomForestClassifier(random_state=42))
    ])
    
    pipe_svm = Pipeline([
        ('scaler', StandardScaler()),
        ('clf', SVC(probability=True, random_state=42))
    ])

    # Define hyperparameter grids to search
    param_grid_rf = {
        'clf__n_estimators': [50, 100, 200],
        'clf__max_depth': [5, 10, 20, None],
        'clf__min_samples_leaf': [1, 2, 4]
    }
    
    param_grid_svm = {
        'clf__C': [0.1, 1, 10, 100],
        'clf__gamma': ['scale', 'auto', 0.01, 0.1],
        'clf__kernel': ['rbf', 'poly']
    }

    # --- 5. Run GridSearchCV for each model ---
    print("\nStarting GridSearchCV for RandomForest...")
    grid_rf = GridSearchCV(pipe_rf, param_grid=param_grid_rf, cv=gkf, n_jobs=-1, scoring='roc_auc', verbose=1)
    grid_rf.fit(X_tr, y_tr, groups=groups_tr)
    best_rf = grid_rf.best_estimator_
    print(f"Best CV score (AUC) for RandomForest: {grid_rf.best_score_:.4f}")
    print(f"Best RandomForest params: {grid_rf.best_params_}")

    print("\nStarting GridSearchCV for SVM...")
    grid_svm = GridSearchCV(pipe_svm, param_grid=param_grid_svm, cv=gkf, n_jobs=-1, scoring='roc_auc', verbose=1)
    grid_svm.fit(X_tr, y_tr, groups=groups_tr)
    best_svm = grid_svm.best_estimator_
    print(f"Best CV score (AUC) for SVM: {grid_svm.best_score_:.4f}")
    print(f"Best SVM params: {grid_svm.best_params_}")
    
    # --- 6. Evaluate Best Models on the Held-Out Test Set ---
    print("\n--- Final Evaluation on Unseen Test Data ---")
    
    print("\n--- Results for Best RandomForest ---")
    y_pred_rf = best_rf.predict(X_te)
    y_prob_rf = best_rf.predict_proba(X_te)[:, 1]
    print(f"Test Set AUC: {roc_auc_score(y_te, y_prob_rf):.4f}")
    print("Confusion Matrix:\n", confusion_matrix(y_te, y_pred_rf))
    print(classification_report(y_te, y_pred_rf))
    
    print("\n--- Results for Best SVM ---")
    y_pred_svm = best_svm.predict(X_te)
    y_prob_svm = best_svm.predict_proba(X_te)[:, 1]
    print(f"Test Set AUC: {roc_auc_score(y_te, y_prob_svm):.4f}")
    print("Confusion Matrix:\n", confusion_matrix(y_te, y_pred_svm))
    print(classification_report(y_te, y_pred_svm))

if __name__ == "__main__":
    main()
```
